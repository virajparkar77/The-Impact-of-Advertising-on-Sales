```python
 from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(X1,y1, test_size=0.3, random_state=42)

X1_train = X1_train.values.reshape(-1, 1)
X1_test = X1_test.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression

lm= LinearRegression()
lm.fit(X1_train, y1_train)
```

```python
predict_Sales=lm.predict(X1_test)
```

```python
sns.displot(predict_Sales - y1_test, kind='kde')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error
import math
```

```python
print('MSE: ', mean_squared_error(y1_test, predict_Sales))
print('MAE: ', mean_absolute_error(y1_test, predict_Sales))
print("RMSE: ", math.sqrt(mean_squared_error(y1_test, predict_Sales)))
print("MedAE: ", median_absolute_error(y1_test, predict_Sales))
```
